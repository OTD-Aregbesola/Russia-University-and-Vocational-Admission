## Russia University and Vocational Admission

This notebook provides information on  University and vocational schools admission in Russia from 2014-2023. The data is gotten from Kaggle.

```python
# libraries for the analysis
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Data Preprocessing

```python
#loading the data
df = pd.read_csv('University and vocational schools admission in Russia 2014-2023.csv', sep=';')
```

```python
df.head(5)
```

```python
df.info()
```

```python
# removing null values
df.dropna(inplace=True)
```

```python
#describe the data
df.describe()
```

```python
#drop columns
df.drop(['Index', 'ID'], axis=1, inplace=True)
```

```python
df.duplicated().sum()
```

#Univariate Analysis

```python
#Visualizing Education level
sns.countplot(x='Education level', data=df)
plt.title('Education Level')
plt.ylabel('Count')
plt.show()
```

From the visual above, we can see that majority have Higer compare to Vocational Education in the Education Level

```python
# Group and count the data
degree_counts = df['Degree'].value_counts()

# Plot the bar chart
plt.barh(degree_counts.index, degree_counts.values, color='skyblue', edgecolor='black')
plt.title('Degree Distribution')
plt.xlabel('Frequency')
plt.ylabel('Degree')
plt.show()
```

```python
#Visual for tution fees
plt.pie(df["Tuition fees"].value_counts(), labels=df["Tuition fees"].value_counts().index, autopct="%1.1f%%", startangle=90)
plt.title('Tuition Fees Distribution')
plt.show()
```

```python
BS_counts = df['Branches of Science'].value_counts()

# Plot the bar chart
plt.barh(BS_counts.index, BS_counts.values, color='skyblue', edgecolor='black')
plt.title('Branches of Science Distribution')
plt.xlabel('Frequency')
plt.ylabel('Braches of Science')
plt.show()
```

#Bivariate Analysis

```python
# The 5 top highest Number of Applicantions for Scholarship/Tuition free Applicant
sorted_df = df.sort_values('Number of Applications', ascending=False)

# Select the top 10 rows
top_5= sorted_df.head(5)

# Display the top 10 rows
display(top_5)
```

```python
#Filtered for Full pay / Tuition fee
df_filtered = df[df['Tuition fees'] == 'Full pay / tuition fee']

# The 5 top highest Number of Applicantions for Full pay / tuition fee.
sorted_df_filtered = df_filtered.sort_values('Number of Applications', ascending=False)

# Select the top 5 rows
top5 = sorted_df_filtered.head(5)

# Display the top 5 rows
display(top5)
```
